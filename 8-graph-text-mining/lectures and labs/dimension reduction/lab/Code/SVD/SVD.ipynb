```python
%matplotlib inline
from numpy import *
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib import rc
import numpy as np

# Load the "gatlin" image data
X = loadtxt('gatlin.csv', delimiter=',')
```

```python
u,s,v = np.linalg.svd(X)
s_diag = diag(s)
plt.figure(1)
plt.imshow(X,cmap = cm.Greys_r)
plt.title('Original image (rank 480)')
plt.axis('off')
plt.show()
```

```python
k=10
X_new = np.dot(u[:,:k],s[:k,:k]).dot(v[:k,:])
plt.figure()
plt.imshow(X_new,cmap = cm.Greys_r)
plt.title('Original image (rank 480)')
plt.axis('off')
```

```python
error = []
for k in [10, 20, 50, 100, 200]:
    X_new = np.dot(u[:,:k],s[:k,:k]).dot(v[:k,:])
    err = np.linalg.norm(X-X_new)/np.linalg.norm(X)
    error.append(err)
plt.plot([10, 20, 50, 100, 200],error)
```
