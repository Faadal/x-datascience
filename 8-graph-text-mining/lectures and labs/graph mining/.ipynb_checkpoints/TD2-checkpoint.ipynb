```python
%matplotlib inline

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

G=nx.read_edgelist("ca-GrQc.txt", comments='#', delimiter='\t',
nodetype=int, create_using=nx.Graph())
```

```python
print G.number_of_nodes()
```

```python
print G.number_of_edges()
```

```python
print nx.number_connected_components(G)
```

```python
GCC=list(nx.connected_component_subgraphs(G))[0] #return connected components order by size
GCC.number_of_nodes()/G.number_of_nodes()
```

```python
degree_sequence = G.degree().values()
print np.min(degree_sequence)
print np.median(degree_sequence)
```

```python
y=nx.degree_histogram(G)
plt.plot(y,'b-',marker='o')
plt.ylabel("Frequency")
plt.xlabel("Degree")
plt.draw()
plt.show()
```

```python
plt.loglog(y,'b-',marker='o')
plt.ylabel("Frequency")
plt.xlabel("Degree")
plt.draw()
plt.show()
```

```python
t=nx.triangles(G)
print t
```

```python
t_values = sorted(set(t.values()))
t_hist = [t.values().count(x) for x in t_values]
plt.plot(t_hist,'b-',marker='o')
plt.ylabel("Frequency")
plt.xlabel("Degree")
plt.draw()
plt.show()
```

```python
nx.average_clustering(G)
```

```python
# Degree centrality
deg_centrality = nx.degree_centrality(G)
# Eigenvector centrality
eig_centrality = nx.eigenvector_centrality(G)
```

```python
sorted_deg_centrality = sorted(deg_centrality.items())
sorted_eig_centrality = sorted(eig_centrality.items())
deg_data = [b for a,b in sorted_deg_centrality]
eig_data = [b for a,b in sorted_eig_centrality]

from scipy.stats.stats import pearsonr
print "Pearson correlation : ", pearsonr(deg_data, eig_data)
plt.scatter(deg_data, eig_data)
```

```python
generator = np.array([[0.99, 0.26], [0.26, 0.53]])
adjacency_prob = generator
k=10
for i in range(1, k):
    adjacency_prob = np.kron(adjacency_prob, generator)
    
import random as rd
adjacency_matrix = np.zeros(adjacency_prob.shape)
for i in range(adjacency_matrix.shape[0]):
    for j in range(i, adjacency_matrix.shape[1]):
        r = rd.random()        
        if(r<adjacency_prob[i][j]):
            adjacency_matrix[i][j] = 1.0
kronecker_graph = nx.from_numpy_matrix(adjacency_matrix, create_using=nx.Graph())
```
