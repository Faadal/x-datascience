```python
%pylab inline
pylab.rcParams['figure.figsize'] = (5, 5)
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_node("Jim")
G.add_node("Jenny")
G.add_node("David")
G.add_edge("Jim", "David")
G.add_edge("Jenny", "David")
print G.number_of_edges()
print G.number_of_nodes()
print G.nodes()
print G.edges()
print G.degree("Jim")
print G.degree()
print nx.number_connected_components(G)
```

```python
G = nx.read_edgelist ("ca-GrQc.txt", comments='#', delimiter='\t', nodetype=int)
print G.number_of_nodes()
print G.number_of_edges()
print nx.number_connected_components(G)
subgraphs = list(nx.connected_component_subgraphs(G))
iMax = np.argmax([g.number_of_nodes() for g in subgraphs])
maxSubgraph = subgraphs[iMax]
print maxSubgraph.number_of_nodes()
print maxSubgraph.number_of_edges()
```

```python
degree_sequence = G.degree().values()
print np.max(degree_sequence)
print np.min(degree_sequence)
print np.mean(degree_sequence)
y=nx.degree_histogram(G)
plt.ylabel("Frequency")
plt.xlabel("Degree")
plt.plot(y,'b-',marker='o')
```

```python
plt.loglog(y)
```

```python
t = nx.triangles(G)
numberOfTriangles = sum(t.values())/3
print numberOfTriangles
```

```python
t_values = sorted(set(t.values()))
t_hist = [t.values().count(x) for x in t_values]
plt.loglog(t_values, t_hist)
```

```python
print nx.average_clustering(G)
deg_centrality = nx.degree_centrality(G)
eig_centrality = nx.eigenvector_centrality(G)
```

```python
sorted_deg_centrality = sorted(deg_centrality.items())
sorted_eig_centrality = sorted(eig_centrality.items())
deg_data = [b for a,b in sorted_deg_centrality]
eig_data = [b for a,b in sorted_eig_centrality]

from scipy.stats.stats import pearsonr
print "Pearson correlation : ", pearsonr(deg_data, eig_data)
plt.scatter(deg_data, eig_data)
```
